```
# data visulization
# importing the libraries
import pandas as pd
import matplotlib.pyplot as plt

# importing the dataset
df = pd.read_csv('Salary_Data.csv')
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values

# bar graph
df.plot(kind='bar')
plt.show()

# histogram
df.plot(kind='hist')
plt.show()

# pie chart
plt.pie(x)
plt.show()

# line graph
plt.plot(x,y,color='red')
plt.show()

# scatter plot
plt.scatter(x,y,color='green')
plt.show()
```
